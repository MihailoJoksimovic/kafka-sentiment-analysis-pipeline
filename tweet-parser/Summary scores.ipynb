```python
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pickle
```

```python
stemming_scores_file = open('stemming_scores.pkl', 'rb')

stemming_scores = np.array(pickle.load(stemming_scores_file))
```

```python
regular_scores_file = open('regular_scores.pkl', 'rb')

regular_scores = np.array(pickle.load(regular_scores_file))
```

```python
lemmatization_scores_file = open('lemmatization_scores.pkl', 'rb')

lemmatization_scores = np.array(pickle.load(lemmatization_scores_file))
```

```python
sns.set_style('ticks')
fig, ax = plt.subplots()
# the size of A4 paper
fig.set_size_inches(11.7, 8.27)
plt.plot(stemming_scores[:, 0], stemming_scores[:, 2], label="Stemming Score")
plt.plot(lemmatization_scores[:, 0], lemmatization_scores[:, 2], label="Lemmatization Score")
plt.plot(regular_scores[10:100, 0], regular_scores[10:100, 2], label="Regular Scores")
plt.xlabel('Number of features (x2)')
plt.ylabel('Accuracy');
ax.legend()
```
